## N1.E01: Calculadora del √≠ndice de masa corporal

Escribe una funci√≥n que calcule el IMC ingresado por el usuario/a, es decir, quien lo ejecute deber√° ingresar estos datos. Puedes obtener m√°s informaci√≥n de su c√°lculo en: https://muysalud.com/salud/indice-de-masa-corporal-imc-que-es-y-como-se-calcula/

La funci√≥n debe clasificar el resultado en sus respectivas categor√≠as.

Consejo: Intenta validar los datos previamente, para que env√≠e un mensaje de advertencia si los datos introducidos por el usuario/a no est√°n en el formato adecuado o no toma valores razonables.

### N1.E01.1 Funciones auxiliares

```python
def entra_valida_numero(mensaje: str) -> float:
    """
    Docstring for entra_valida_numero
    
    La funci√≥n te pide que introduzcas un n√∫mero mostrandote el mensaje de que numero pide. Despues lo valida en este orden:
    
    1.- Que no usas el separador decimal coma (,) y si lo usas te avisa y te vuelve a padir el n√∫mero
    
    2.- Que el n√∫mero introducido sea con caracteres numericos, que no haya ning√∫n texto
    
    3.- Que el n√∫mero es mayor que cero (ni cero ni negativo)
    
    Si alguna validaci√≥n no se cumple, te lo indica y te vuelve a pedir el n√∫mero
    
    :param mensaje: Mensaje del n√∫mero que se pide
    :type mensaje: str
    :return: El numero convertido a tipo float
    :rtype: float
    """
    while True:
        entrada = input(mensaje)

        # No permitir coma como separador decimal
        if "," in entrada:
            print("Usa punto (.) como separador decimal.")
            continue

        # Intentar convertir a float
        try:
            numero = float(entrada)

        except ValueError:
            print("Debes introducir un n√∫mero v√°lido, no texto.")

            continue

        # Validar que sea mayor que cero
        if numero <= 0:
            print("El n√∫mero debe ser mayor que cero.")
            continue

        return numero
```

```python
def calculo_imc (peso_imc:float, altura_imc:float) -> float:
    """
    Docstring for calculo_imc
    
    Calcula el Indice de Masa Corporal (IMC) apartir del peso y la altura aplicando la formula peso / (altura^2)
    
    :param peso: Peso en Kilogramos
    :type peso: float
    :param altura: Altura en metros
    :type altura: float
    :return: Valor de IMC
    :rtype: float
    """
    
    return round(peso_imc / (altura_imc**2),2)
```

```python
def clasifica_imc(imc_valor:float)->str:
    """
    Docstring for clasifica_imc 
    
    Recibe le valor numerico del IMC y lo clasifica segun los rangos:
    Bajo peso: menos de 18,5.
    Peso normal: 18,5-24,9.
    Sobrepeso: 25-29,9.
    Obesidad: m√°s de 30.
    
    :param imc_valor: Es el valor numerico del imc
    :type imc_valor: float
    :return: La categoria asignada seg√∫n su valor numerico de imc
    :rtype: str
    """
    if imc_valor < 18.5:
        return "Bajo peso"
    elif (imc_valor >= 18.5) and (imc_valor <= 24.9):
        return "Peso normal"
    elif 25 <= imc_valor <= 29.9:
        return "Sobrepeso"
    else:
        return "Obesidad"
```

```python
def ejecuta_calculadora_imc():
    """Ejecuta paso por paso desde la entrada de datos hasta la salido del imc y su clasificaci√≥n
    """
    #Entrada de peso y altura
    peso = entra_valida_numero("Escribe tu peso en Kgr (usa punto para decimales)")
    altura = entra_valida_numero("Escribe tu altura en Metros (usa punto para decimales)")

    # considero un warning cuando el peso es superior a 250 Kgrs. y/o la altura es superior a 2.50 metros
    if peso > 250:
        print(f"ADVERTENCIA: Valor de peso ({peso} Kgrs.) no es razonable")
        
    if altura > 2.50:
        print(f"ADVERTENCIA: Valor de altura ({altura} metros) no es razonable")

    #Calculo del imc
    imc = calculo_imc(peso, altura)

    #Clasificaci√≥n del imc
    rango = clasifica_imc(imc)

    #Imprimir resultados
    print(f"\nPara : {peso} Kgr de peso y {altura} metros de altura \nTienes un IMC de {imc} que corresponde a {rango}")
```

### N1-E01.2 Programa principal

```python
ejecuta_calculadora_imc()
```

## N1.E02: Conversor de temperaturas

Existen diversas unidades de temperatura utilizadas en distintos contextos y regiones. Las m√°s comunes son Celsius (¬∞C), Fahrenheit (¬∞F) y Kelvin (K). Tambi√©n existen otras unidades como Rankine (¬∞Ra) y R√©aumur (¬∞Re).

Selecciona al menos 2 conversores, de modo que al introducir una temperatura devuelva, como m√≠nimo, dos conversiones, de modo que se puedan guardar (recuerda que un print() no se puede guardar nunca).

Consejo: Intenta validar los datos previamente, para que env√≠e un mensaje de advertencia si los datos introducidos por el usuario/a no est√°n en el formato adecuado.

(EXTRA): Piensa una manera de almacenar todas las posibles conversiones en un solo objeto (¬øLista? ¬øDiccionario? ¬øDataFrame?) en vez de escribir muchos if else en funci√≥n de la temperatura de origen y la temperatura de destino.

Web calculadora para comprobar: https://www.mouser.es/technical-resources/conversion-calculators/temperature-conversion-calculator?srsltid=AfmBOoqCzjN4_JSu9a14aGbRuxu5AvjEFnlNx3zFF_sokR5QZYWA9bGF

### N1.E02.1 Funciones auxiliares

```python
def entra_valida_grados() -> float:
    """
    Docstring for entra_valida_numero
    
    La funci√≥n te pide que introduzcas un n√∫mero mostrandote el mensaje de que numero pide. Despues lo valida en este orden:
    
    1.- Que no usas el separador decimal coma (,) y si lo usas te avisa y te vuelve a padir el n√∫mero
    
    2.- Que el n√∫mero introducido sea con caracteres numericos, que no haya ning√∫n texto
    
    Si alguna validaci√≥n no se cumple, te lo indica y te vuelve a pedir el n√∫mero
    
    :return: El numero convertido a tipo float
    :rtype: float
    """
    while True:
        entrada = input("Introduce los grados de temperatura originales a convertir")

        # No permitir coma como separador decimal
        if "," in entrada:
            print("Usa punto (.) como separador decimal.")
            continue

        # Intentar convertir a float
        try:
            numero = float(entrada)

        except ValueError:
            print("Debes introducir un n√∫mero v√°lido, no texto.")

            continue

        return numero
```

```python
def entra_valida_tipo_temp()->str:
    """
    Docstring for entra_valida_tipo_temp
    
    Pide la entrada del tipo de grados conocido o inicial para trasformar a los otros 4 tipos, seg√∫n:
    (C) Celsius    
    (F) Fahrenheit     
    (K) Kelvin    
    (RE) Reaumur    
    (RA) Rankine
    
    :return: Devuelve la inicial 
    :rtype: str
    """
    
    tupla_temp = ("c","C","f","F","k","K","re","RE","Re","rE","ra","RA","Ra","rA")
    men = """
    Elige una opci√≥n del tipo de grados originales a convertir:  
    (C) Celsius    
    (F) Fahrenheit     
    (K) Kelvin    
    (RE) Reaumur    
    (RA) Rankine
    """
    
    while True:
        entrada = input(men).strip().lower()
        
        if entrada not in tupla_temp:
            print("Error al introducir el tipo de grados de temperatura. Vuelve a intentarlo")
            continue
        
        return entrada
```

```python
def convertir(tipo:str, temp:float)->dict:
    """
    Docstring for convertir
    
    :param tipo: Es el tipo de grados iniciales a partir de los cuales hare la conversion al resto de tipos: "C", "F", "K", "RE" Y "RA"
    :type tipo: str
    :param temp: Son los grados iniciales que los convertira a los grados en el resto de tipos
    :type temp: float
    :return: Devuelve un diccionario con la equivalencia de los grados en los 5 tipos de medidas diferentes
    :rtype: dict
    """

    salida = {}
    
    # Conversi√≥n a Celsius desde cualquier tipo
    match  tipo:
        case "f":
            salida["c"] = round((temp - 32) / 1.8,2)
        case "k":
            salida["c"] = round(temp - 273.15,2)
        case "re":
            salida["c"] = round(temp / 0.8,2)
        case "ra":
            salida["c"] = round((temp - 491.67) / 1.8,2)
        case _:
            salida["c"] = round(temp,2)
    
    # Convertir desde Celsius a todos los tipos  
    salida["f"] = round(salida["c"] * 1.8 + 32,2)
    salida["k"] = round(salida["c"] + 273.15,2)
    salida["re"] = round(salida["c"] * 0.8,2)
    salida["ra"] = round(salida["c"] * 1.8 + 491.67,2)
    
    return salida
```

```python
def ejecuta_conversor_temp():
    """Ejecuta paso por paso la conversion a los 5 tipos de grados a partir de la entrada de uno de ellos
    """
    entrada_temp = entra_valida_grados()
    entrada_tipo = entra_valida_tipo_temp()
    conversion = (convertir(entrada_tipo, entrada_temp))

    prefijo = {
        "c": "Celsius", 
        "f": "Fahrenheit", 
        "k": "Kelvin", 
        "re": "Reaumur", 
        "ra": "Rankine" 
        }

    sufijo = {
        "c": "¬∞C", 
        "f": "¬∞F", 
        "k": "K", 
        "re": "¬∞Re", 
        "ra": "¬∞Ra" 
        }

    print(f"{'TIPO':<10}   {'GRADOS':>20}")
    for clave, valor in conversion.items():
        print(f"{prefijo[clave]:<10} : {valor:>20} {sufijo[clave]} ")
```

```python
ejecuta_conversor_temp()
```

## N1.E03: Contador de palabras de un texto.

Escribe una funci√≥n que dado un texto, muestre las veces que aparece cada palabra. Intenta que se gestionen todas las posibles casu√≠sticas que hagan que el programa no funcione correctamente.

(EXTRA): ¬øCu√°l es la longitud media de las palabras del texto que has escrito? ‚Äú¬øHola c√≥mo va?‚Äù deber√≠a devolver (4+3+2) / 3 = 3

### N1.E03.1 Funciones auxiliares

```python
def entra_texto()->str:
    """Pide entrar el texto que queremos contar el n√∫mero de veces que aparece cada palabra y luego la longitud media de las palabras introducidas

    Returns:
        str: Es el texto tal y como se ha introducido
    """
    texto_original = input("Entra el texto a contar palabras y calcular la longitud media de las palabras")
    return texto_original
```

```python
def limpia_texto (texto_sucio:str)->str:
    """Convierte el texto que le llega todo a minusculas y elimina los retornos de carro, los caracteres gramaticales y los numeros

    Args:
        texto_sucio (str): Recibe el texto a limpiar

    Returns:
        str: Devuelve el texto limpio ya tratado
    """
    
    caracteres = set(".,:;-_()[]{}<>¬ø?¬°!/&%$‚Ç¨#@|\'*=+^1234567890")  #conjunto de caracteres a eliminar
    
    texto_limpio_final = ""
    texto_limpio =texto_sucio.lower()   #Lo conviero todo a minusculas para no hacer distincion entre palabras en minusculas y mayusculas
    texto_limpio = texto_limpio.replace("\n"," " )  #Elimino los retornos de carro por si el texto tiene varias l√≠neas
    
    for letra in texto_limpio:
        if letra not in caracteres:
            texto_limpio_final += letra
        
    return texto_limpio_final.strip()   #Elimino los espacios por la izquierda y la derecha del texto
    
```

```python
def cuenta (texto_a_contar:str)->dict:
    """Recibe un texto y devuelve las veces que aparece cada palabra en el texto

    Args:
        texto_a_contar (str): Texto a contar las apariciones de cada palabra

    Returns:
        dict: Diccionario con la palabra como clave y el n√∫mero de repeticiones como valor
    """
    
    cuenta_palabras = {}
    texto_a_contar = texto_a_contar.split()
    
    for palabra in texto_a_contar:
        if palabra in cuenta_palabras:
            cuenta_palabras[palabra] += 1
        else:
            cuenta_palabras[palabra] =1

    return cuenta_palabras
```

```python
def longitud_media_palabra(cuenta_palabras:dict):
    """Cuenta el n√∫mero de palabras teniendo en cuenta si se repeiten y el numero de letras para calcular la longitud media de las palabras

    Args:
        cuenta_palabras (dict): Diccionario con clave la palabra del texto y valor sus repeticiones en el texto
    """

    n_veces = 0
    suma_letras = 0
    for clave, valor in cuenta_palabras.items():
        n_veces += valor
        suma_letras += len(clave) * valor

    print(f"N√∫mero total de palabras: {n_veces}")
    print(f"Letras totales en el texto: {suma_letras}")
    
    try:
        print(f"Longitud media de las palabras del texto es: {round(suma_letras/n_veces,2)} letras")
    except ZeroDivisionError:
        print(f"Debes haber introducido al menos una palabra para poder hacer una media")
```

```python
def ejecuta_cuenta_palabras():
    """Ejecuta paso por paso el recuento de palabras y su longitud media de un texto que te pide introducir
    """
    texto = entra_texto()
    print(f"Texto original:  '{texto}'\n")

    texto_sin_caracteres = limpia_texto(texto)

    palabras = cuenta(texto_sin_caracteres)
    print(f"{'PALABRA':<10} : {'APARICIONES':>5}")
    for clave, valor in palabras.items():
        print(f"{clave:<10} : {valor:>5}")

    print("\n\n")
    longitud_media_palabra(palabras)
```

### N1.E03.2 Programa principal

```python
ejecuta_cuenta_palabras()
```

## N1.E04: Diccionario inverso (con posibilidad de duplicados).

Resulta que el cliente tiene una encuesta muy antigua que se almacena en un diccionario y los resultados los necesita a la inversa, es decir, intercambiando las claves y valores. Los valores y claves en el diccionario original son √∫nicos; si √©ste no es el caso, la funci√≥n deber√≠a imprimir un mensaje de advertencia, junto con una lista con los valores asociados a la clave repetida.


### N1.E04.1 Funciones auxiliares

```python
def invertir(dic_original:dict):

    dic_inverso ={}
    advertencia = False

    for clave, valor in dic_original.items():
        if valor in dic_inverso:
            dic_inverso[valor].append(clave)
            advertencia = True
        else:
            dic_inverso[valor] = [clave]
    return dic_inverso, advertencia

```

```python
def printa_diccionario(dic_a_printar:dict, adver:bool,original:bool):
    if adver:
        print("\n**************************************************************************************************************")
        print("** ADVERTENCIA: hay valores originales repetidos que hacen tener claves repetidas en en diccionario inverso **")
        print("**************************************************************************************************************")
    if original:
        print("diccionario original\n")
        print(f"{'CLAVE ORIGINAL':<15} : {'VALOR ORIGINAL':>5}")
    else:
        print("\ndiccionario invertido\n")
        print(f"{'NUEVA CLAVE':<15} : {'NUEVO VALOR/es':>5}")
        
    for clave, valor in dic_a_printar.items():
        print(f"{clave:<15} : {valor}")
```

```python
def ejecuta_dict_inverso(dic_original):
    """Ejecuta paso por paso el diccionario inverso al que recibe dando una advertencia si hay valores repetidos en el diccionario original

    Args:
        dic_original (_type_): Diccionario original que queremos su inverso
    """

    #Imprimo el diccionario original
    printa_diccionario(dic_original,False,True)

    #Invierto el diccionario
    dic_invertido, advertencia = invertir(dic_original)

    #Imprimo el diccionario invertido
    printa_diccionario(dic_invertido,advertencia,False)
```

### N1-E04.2 Programa principal

```python
dic_inicial ={
    "a":1,
    "b":2,
    "c":2,
    "d":4,
    "e":1
}

ejecuta_dict_inverso(dic_inicial)
```

## N2.E01: Contador y ordenador de palabras de un texto.

El cliente qued√≥ contento con el contador de palabras, pero ahora quiere leer archivos TXT y que calcule la frecuencia de cada palabra ordenadas dentro de las entradas habituales del diccionario seg√∫n la letra con la que comienzan, es decir, las claves deben ir de la A a la Z y dentro de la A debemos ir de la A a Z.

### N2.E01.1 Funciones auxiliares

```python
"""Importar todas las librerias:
- string para poder utilizar la funcion que imprime un diccionario m√°s visual
"""

from pprint import pprint
```

```python
def leer_archivo(file_name:str)->str:
    """Abre el fichero que le paso, con su path, lee su contenido pasandolo a una variable str y cierra el fichero

    Args:
        file_name (str): Nombre del fichero tipo txt con el texto a tratar

    Returns:
        str: Contenido del texto del fichero .txt
    """
    
    # Abro el fichero en modo lectura  
    try:
        file = open(file_name, "r")
        # leo el contenido del fichero y lo paso a tipo str
        texto = file.read()
        # Cierro el fichero
        file.close()
        return texto
        
    except FileNotFoundError:
        print(f"ERROR. El fichero '{file_name}' no se ha encontrado")
        return None
    
```

```python
def agrupa_alfabeticamente(dic_sin_agrupar:dict)->dict:
    """Entra un diccionario, lo ordena alfabeticamente por la clave y devuelve un diccionario agrupado alfabeticamente por la clave, 
    siendo los valores de cada clave otro diccionario ordenado alfabeticamente

    Args:
        dic_sin_agrupar (dict): Diccionario sin agrupar

    Returns:
        dict: Diccionario ordenado alfabeticamente y agrupado, siendo los valores, diccionarios ordenados alfabeticamente tambien
    """

    dic_agrupado = {}
    
    for clave, valor in sorted(dic_sin_agrupar.items()):
        letra =clave[0].lower()  # obtengo la primera letra de cada clave
        if letra not in dic_agrupado:
            dic_agrupado[letra] = {}   #Creo un diccionario vacio con la primera letra
        
        dic_agrupado[letra][clave] = valor  # pongo la clave y su valor dentro de la clave letra y como ya esta ordenado no hace falta volver a ordenarlo
    
    
    return dic_agrupado
```

```python
def ejecuta_cuenta_ordena_palabras(archivo:str):
    """Ejecuta paso por paso el recuento de palabras y las ordena alfabeticamente en un fichero txt

    Args:
        archivo (str): fichero txt que contine las palabras
    """
    # Leer el texto contenido en el fichero que esta en el directorio de trabajo y obtengo un str
    texto_fichero = leer_archivo(archivo)
    if texto_fichero is None:
        print("Fin de programa. No se puede continuar sin un fichero con los datos")
    else:
        #Uso la funci√≥n de limpiar texto definida en el ejercicio 3 del Nivel 1
        texto_fichero_sin_caracteres = limpia_texto(texto_fichero)

        #Uso la funci√≥n de contar palabras en el texto definida en el ejercicio 3 del Nivel 1 y las ordeno alfabeticamente
        palabras = cuenta(texto_fichero_sin_caracteres)

        palabras_agrupadas = agrupa_alfabeticamente(palabras)
        pprint(palabras_agrupadas)
```

## N2.E02: Conversi√≥n de tipos de datos.

El cliente recibe una lista de datos y necesita generar dos listas, la primera donde estar√°n todos los elementos que pudieron convertirse en flotantes y la otra donde est√°n los elementos que no pudieron convertirse. 

### N2.E02.1 Funciones auxiliares

Como cada objeto puede contener a su vez otros objetos y no sabemos el grado de profundidad que podemos alcanzar (lista que incluye diccionarios que a su vez incluyen otros objetos y asi un nuemro indeterminado de veces). Para resolver estos problemas es tipico hacerlo usando la ***recursividad*** (al igual que el t√≠pico calculo del factorial de un n√∫mero), llamando cada iteracion a si mismo hasta que se acaban los elementos que estamos iterando



Para clasificarlos hay que tener en cuenta que convertir a float() **NO funciona dando un error** con:

    - listas ‚Üí float([1,2]) ‚Üí TypeError

    - tuplas ‚Üí float((1,2)) ‚Üí TypeError

    - sets ‚Üí float({1,2}) ‚Üí TypeError

    - diccionarios ‚Üí float({"a":1}) ‚Üí TypeError

    - strings no num√©ricos ‚Üí float("hola") ‚Üí ValueError

```python
def clasificar_floats(obj: any, lista_float: list, lista_no_float: list):
    """Recorre cualquier estructura (listas, tuplas, sets, diccionarios)
    y clasifica todos los elementos simples en:
    - convertibles a float
    - no convertibles

    Args:
        obj (any): Objeto a evaluar que puede ser desde simple a tupla, lista, set o diccionario
        lista_float (list): Se almacenan los objetos simples que se han podido convertir a float. Variable global
        lista_no_float (list): Se almacenan los objetos simples que NO se han podido convertir a float. Variable global

    Returns:
        _type_: En la recursividad, solo sirven para detener la ejecuci√≥n en ese caso concreto, no para devolver valores.
    """


    # Paso 1: diccionario ‚Üí recorrer valores
    if isinstance(obj, dict):
        for v in obj.values():
            clasificar_floats(v, lista_float, lista_no_float)
        return  # Es para indicar que ya se ha procesado todo el diccionario y corta la ejecucion

    # Paso 2: lista, tupla o set ‚Üí recorrer elementos
    if isinstance(obj, (list, tuple, set)):
        for elem in obj:
            clasificar_floats(elem, lista_float, lista_no_float)
        return   # Es para indicar que ya se ha procesado toda la lista, tupla o set y corta la ejecucion

    # Paso 3: valor simple ‚Üí intentar convertir y clasificarlo
    try:
        lista_float.append(float(obj))
    except (ValueError, TypeError):
        lista_no_float.append(obj)
```

```python
def ejecuta_conversion_foat(lista_inicial:list):
    """Ejecuta paso por paso la clsificacion de los datos de una lista en si se pueden o no convertir a tipo float

    Args:
        lista_inicial (list): lista con los datos a clasificar
    """

    lista_float = []
    lista_no_float = []

    clasificar_floats(lista_inicial, lista_float, lista_no_float)

    print(f"lista inicial: {lista_inicial}")
    print("Convertibles a float:", lista_float)
    print("No convertibles a float:", lista_no_float)
```

```python
"""Importar todas las librerias:
- string para tener todos los caracteres sin tener que crear yo una lista de ellos
- numpy para usar sus funciones y m√©todos aleatorios
- piperclip para copiar un string en el portapapeles-
"""
import string
import numpy as np
import pyperclip
```

```python
def entra_parametro(texto:str)->bool:
    """Recibe el texto con el nombre del par√°metro a preguntar si quieres que este presente.
    Comprueba que la respuesta es si o no, y si no lo es te avisa y te vuelve a preguntar hasta que introduzcas una respuesta correcta.
    Devuelve una variable booleam con True para si y con False para no

    Args:
        texto (str): Nombre del par√°metro que aparece en la pregunta

    Returns:
        bool: Respuesta afirmativa o negativa a la pregunta si quieres que este presente este tipo de par√°metro cuyo nombre has pasado
    """
    resp_positiva = ["si", "s√≠"]
    resp = ["si", "s√≠","no"]
    while True:
        entrada = input(f"Quieres que el psw contenga {texto} (SI / NO)").strip().lower()

        # Validar que sea una respuesta dentro de las posibles
        if entrada not in resp:
            print("Respuesta erronea. Debes contestar Si o NO")
            continue
        else:
            if entrada in resp_positiva:
                respuesta = True
            else:
                respuesta = False

        return respuesta
```

```python
def entra_parametros()->list:
    """Contiene una relacion de los par√°metros que quieres preguntar para preguntar individualmente por cada uno de ellos.
    Devolver√° una lista con las respuestas True o False para cada par√°metro

    Returns:
        list: Lista con cada elemento es la respuesta a cada par√°metro preguntado
    """
    mayus = entra_parametro("Mayusculas")
    minus = entra_parametro("Minusculas")
    num = entra_parametro("Numeros")
    caract = entra_parametro("Simbolos")
    
    return [mayus, minus, num, caract]
```

```python
def entra_longitud() -> int:
    """Pide la entrada de un numetro entero mayor que cero para la longitud del psw. El bucle finaliza cuando se entra un n√∫mero v√°lido

    Returns:
        int: Devuelve la longitud del psw
    """
    
    while True:
        entrada = input("Entra la longitud del psw que deseas (positivo y mayor que cero)")

        # Intentar convertir a float
        try:
            numero = int(entrada)

        except ValueError:
            print("Debes introducir un n√∫mero valido:ni decimal ni texto.")

            continue

        # Validar que sea mayor que cero
        if numero <= 0:
            print("El n√∫mero debe ser mayor que cero.")
            continue

        return numero
```

```python
def verificacion(longitud:int, mayusculas:bool, minusculas:bool, numeros:bool, signos:bool)-> bool:
    """Recibe la longitud y cada parametro si hay o no que incluirlo en el psw y verifica:
    1.- Que haya por lo menos un par√°metro elegido (que este a True). Si no hay ninguno imprime mensaje comunicandolo
    2.- Que la longitud pedida sea mayor o igual a la suma de tipos de par√°metros solicitados. Si no lo es, imprime mensaje comunicandolo
    Solo si las 2 condiciones se cumplen, devuelve un resultado boolean verdadero, en caso contrario lo devuelve falso

    Args:
        longitud (int): longitud del psw que pides
        mayusculas (bool): Si quieres o no que incluya mayusculas
        minusculas (bool): Si quieres o no que incluya minusculas
        numeros (bool): Si quieres o no que incluya n√∫meros
        signos (bool): Si quieres o no que incluya s√≠mbolos

    Returns:
        bool: Resultado de la verificaci√≥n
    """
    
    min_longitud = sum([mayusculas, minusculas, numeros, signos])
    
    if min_longitud == 0:
        verificacion = False
        print("No has elegido ning√∫n tipo de par√°metro. Debes elegir por lo menos uno entre: May√∫sculas, min√∫sculas, n√∫meros y signos")
    else:    
        if (min_longitud <= longitud):
            verificacion = True
        else:
            verificacion = False
            print(f"la longitud introdicida ({longitud}) debe ser mayor o igual a la suma de par√°metros elegidos ({min_longitud})")
        
    return verificacion
    
```

```python
def crear_psw (longitud:int, mayusculas:bool, minusculas:bool, numeros:bool, signos:bool)->str:
    """Crea un psw con la longitud pedida e incluyendo los typos de caracteres pedidos

    Args:
        longitud (int): Logitud del psw
        mayusculas (bool): Que incluya o no mayusculas
        minusculas (bool): Que incluya o no minusculas
        numeros (bool): Que incluya o no n√∫meros
        signos (bool): Que incluya o no s√≠mbolos

    Returns:
        str: PSW creado al azar
    """
    
    letras_minus = list(string.ascii_lowercase)
    letras_mayus = list(string.ascii_uppercase)
    digitos = list(string.digits)
    simbolos = list(string.punctuation)
    todos = []
    psw_formandose = []
    cantidad_tipos =0
    
    # Paso 1: Elegimos 1 caracter obligatorio por tipo pedido (es el minimo obligatorio). 
    #          Lo a√±adimos al psw y a√±adimos todos los caracteres del tipo a una lista para luego elegir al azar los que queden hasta la longitud
    if mayusculas:
        cantidad_tipos +=1
        mayus = np.random.choice(np.array(list(string.ascii_uppercase)))
        psw_formandose.append(mayus)
        todos.extend(letras_mayus)
    
    if minusculas:
        cantidad_tipos +=1
        minus = np.random.choice(np.array(list(string.ascii_lowercase)))
        psw_formandose.append(minus)
        todos.extend(letras_minus)
    
    if numeros:
        cantidad_tipos +=1
        numer = np.random.choice(np.array(list(string.digits)))
        psw_formandose.append(numer)
        todos.extend(digitos)
    
    if signos:
        cantidad_tipos +=1
        simbol = np.random.choice(np.array(list(string.punctuation)))
        psw_formandose.append(simbol)
        todos.extend(simbolos)
    
    # Paso 2: Elegimos al azar el resto de caracteres hasta llegar a la longitud pedida. 
    #         Usaremos la lista todos que hemos ido creando con los tipos de caracteres pedidos, pero permutandolos para mayor azar
    
    todos_array = np.random.permutation(np.array(todos))
    
    for contador in range(longitud - cantidad_tipos):
        car = np.random.choice(todos_array)
        psw_formandose.append(car)

    
    # Paso 3. mezcalmos el psw obtenido para mayor azar o aleatoriedad
    psw_final = np.random.permutation(np.array(psw_formandose))
    
    # Paso 4, lo convierto a string
    psw_final = "".join(psw_final)
    
    
    return psw_final
    
```

```python
def copiar_portapapeles(psw:str):
    """Copia en el portapapeles el string que recibe.

    Args:
        psw (str): String que pegar√° en el portapapeles.
    """
    pyperclip.copy(psw)
    print("Psw copiado en tu prtapapeles")
```

```python
def ejecuta_genera_psw():
    """Ejecuta paso por paso la generacion de un nuevo psw con la longitud y tipos de caracteres que te pedir√°
    """
    # Paso 1. Entrada de datos
    longi = entra_longitud()
    lista_parametros = entra_parametros()

    # Paso 2. Verificaciones si la longitud es acorde a los tipos de caracteres solicitados y si hay al menos un tipo solicitado
    verificaciones = verificacion(longi, *lista_parametros)

    # Paso 3. Crear el psw
    if verificaciones:
        new_psw = crear_psw(longi, *lista_parametros)
        print(f"La psw generada es {new_psw}")
        
        # Paso 4 copiar al portapapeles
        copiar_portapapeles(new_psw)
```

```python
"""Importar todas las librerias:
- pandas para poder usar los Data Frames
"""
import pandas as pd
```

```python
def importar_txt_df(fichero:str)->pd.DataFrame:
    """Recibe el nombre de un fichero txt √≥ csv con su path si esta fuera del directorio de trabajo y devuelve un DataFrame.
    No usa ninguna fila como cabecera del DF resultante
    Tiene en cuenta los separadores \t, \n y -
    Pone nombre a las columnas del DF: "equipo_A", "goles_A", "goles_B", "equipo_B"

    Args:
        fichero (str): Fichero con todos los partidos de todas las jornadas en formato EquipoA\tgolesA-golesB\tEquipoB

    Returns:
        pd.DataFrame: DF con una columna para cada partido llamadas:"equipo_A", "goles_A", "goles_B", "equipo_B"
    """

    
    try:
        df = pd.read_csv(fichero,
                    header=None,
                    sep=r"[\t-]",
                    names = ["equipo_A", "goles_A", "goles_B", "equipo_B"]
        )
        return(df)
    except FileNotFoundError:
        print(f"ERROR: no se encontr√≥ el fichero '{fichero}'")
        return None
```

```python
def lista_equipos(df:pd.DataFrame)->list:
    """Recibe el DF con todos los partidos y obtiene la lista unica de todos los equipos prticipantes.
    Concatena los equiposA y los equiposB por si hubiese alg√∫n equipo que solo hubiese jugado 1 partido con local o visitante

    Args:
        df (pd.DataFrame): DF con todos los partidos y sus resultados

    Returns:
        list: Lista con el nombre de todos los equipos que participan en la competici√≥n
    """
    return pd.unique(pd.concat([df["equipo_A"], df["equipo_B"]]))
```

```python
def crear_clasificacion(equipos:list)->pd.DataFrame:
    """Crea un nuevo DF que sera la tabla de clasificaci√≥n, cuyo indice ser√°n los equipos que recibe en una lista y 
    las columnas: "partidos_ganados", "partidos_empatados", "partidos_perdidos","goles_favor", "goles_encontra", "puntos".
    Rellena todo del DF con el n√∫mero entero 0 (cero)

    Args:
        equipos (list): lista unica de equipos participantes

    Returns:
        pd.DataFrame: DF con indice los equipos y columnas mencionadas y relleno del numero entero 0
    """
    return pd.DataFrame(
        index = equipos,
        columns=["partidos_ganados", "partidos_empatados", "partidos_perdidos","goles_favor", "goles_encontra", "puntos"]
    ).fillna(0).astype(int)
```

```python
def contar_goles(df_par:pd.DataFrame, df_clasi:pd.DataFrame)-> pd.DataFrame:
    """Recibe el DF con todos los partidos del que sumar√° para cada equipo todos los goles a favor y en contra tanto en partidos como local y 
    como visitante y los pone para cada equipo en la tabla clasificacion que es el otro DF que recibe y luego lo devuelve relleno con estos datos

    Args:
        df_par (pd.DataFrame): DF con todos los partidos jugados
        df_clasi (pd.DataFrame): DF con la tabla de clasificaci√≥n

    Returns:
        pd.DataFrame: DF con la tabla de clasificaci√≥n actualizandos los goles marcados y recibidos
    """

    for fila in df_par.index:
        # Goles a favor
        df_clasi.loc[df_par.loc[fila,"equipo_A"], "goles_favor"] += df_par.loc[fila,"goles_A"]
        df_clasi.loc[df_par.loc[fila,"equipo_B"], "goles_favor"] += df_par.loc[fila,"goles_B"]
        
        # Goles en contra
        df_clasi.loc[df_par.loc[fila,"equipo_A"], "goles_encontra"] += df_par.loc[fila,"goles_B"]
        df_clasi.loc[df_par.loc[fila,"equipo_B"], "goles_encontra"] += df_par.loc[fila,"goles_A"]
        
    df_clasi["dif_goles"] = df_clasi["goles_favor"] - df_clasi["goles_encontra"]
    return df_clasi

```

```python
def contar_reultado_partidos(df_par:pd.DataFrame, df_clasi:pd.DataFrame)-> pd.DataFrame:
    """Recibe el DF con todos los partidos del que sumar√° para cada equipo todos sus partidos ganados, perdidos y empatados, tanto en partidos como local  
    como visitante y los pone para cada equipo en la tabla clasificacion que es el otro DF que recibe y luego lo devuelve relleno con estos datos actualizados

    Args:
        df_par (pd.DataFrame): DF con todos los partidos jugados
        df_clasi (pd.DataFrame): DF con la tabla de clasificaci√≥n

    Returns:
        pd.DataFrame: DF con la tabla de clasificaci√≥n actualizanda con los partidos ganados, perdido y empatados
    """
    #Como local
    for fila in df_par.index:
        if (df_par.loc[fila,"goles_A"] > df_par.loc[fila,"goles_B"]):
            df_clasi.loc[df_par.loc[fila,"equipo_A"], "partidos_ganados"] += 1
        elif df_par.loc[fila,"goles_A"] < df_par.loc[fila,"goles_B"]:
            df_clasi.loc[df_par.loc[fila,"equipo_A"], "partidos_perdidos"] += 1
        else:
            df_clasi.loc[df_par.loc[fila,"equipo_A"], "partidos_empatados"] += 1
            
    #Como Visitante
    for fila in df_par.index:
        if (df_par.loc[fila,"goles_B"] > df_par.loc[fila,"goles_A"]):
            df_clasi.loc[df_par.loc[fila,"equipo_B"], "partidos_ganados"] += 1
        elif df_par.loc[fila,"goles_B"] < df_par.loc[fila,"goles_A"]:
            df_clasi.loc[df_par.loc[fila,"equipo_B"], "partidos_perdidos"] += 1
        else:
            df_clasi.loc[df_par.loc[fila,"equipo_B"], "partidos_empatados"] += 1
    
    return df_clasi
```

```python
def calcular_puntos(df_clasi:pd.DataFrame) -> pd.DataFrame:
    """Recibe el DF con la tabla clasificaci√≥n y lo devuelve con los puntos totales por equipo actualizado, segun:
    partido ganado son 3 puntos, partido empatado son 1 punto y partido perdido son 0 puntos

    Args:
        df_clasi (pd.DataFrame): DF con la tabla de clasificaci√≥n

    Returns:
        pd.DataFrame: DF con la tabla de clasificaci√≥n actualizanda con los puntos de cada equipo
    """
    for fila in df_clasi.index:
        suma_puntos = (df_clasi.loc[fila,"partidos_ganados"] * 3) +(df_clasi.loc[fila,"partidos_empatados"])
        df_clasi.loc[fila,"puntos"] = suma_puntos
        
    return df_clasi
```

```python
def ordenar_clasificacion (df_clasi:pd.DataFrame) -> pd.DataFrame:
    """Recibe el DF con la clasificaci√≥n y lo ordena segun el criterio: por puntos descendente y en caso de mismos puntos por diferencia de goles
    a favor menos goles en contra y tambien descendente.

    Args:
        df_clasi (pd.DataFrame): DF con la tabla de clasificaci√≥n

    Returns:
        pd.DataFrame: DF con la tabla de clasificaci√≥n actualizanda con el orden de la clasificacion de los equipos en la competici√≥n.
    """
    df_clasi = df_clasi.sort_values(by=["puntos", "dif_goles"], ascending=[False, False])
    
    return df_clasi
```

```python
def ejecutar_procesado_liga(partidos_resultados:str):
    """Ejecuta paso por paso el procesamiento desde un fichero con todos los partidos y sus resultados hasta hacer la tabla clasificatoria

    Args:
        partidos_resultados (str): Fichero con todos los partidos y sus resultados
    """
    
    # Paso 1. importar el fichero csv a un Data Frame de pandas
    df_partidos = importar_txt_df(partidos_resultados)
    if df_partidos is None:
        print("Fin de programa. No se puede continuar sin un fichero con datos de los partidos y sus resultados\n")
    else:

        # Paso 2. Crear una lista con los equipos de la competici√≥n
        todos_equipos = lista_equipos(df_partidos)

        # Paso 3. Crear la tabla clasificacion con el indice los equipos y rellena de ceros (integer)
        df_clasificacion = crear_clasificacion(todos_equipos)

        # Paso 4. Calcular los goles a favor
        df_clasificacion = contar_goles(df_partidos,df_clasificacion)

        # Paso 5. Contar partidos con resultado ganados, empatados y perdidos
        df_clasificacion = contar_reultado_partidos(df_partidos,df_clasificacion)

        # Paso 6. Calcular puntos de cada equipo
        df_clasificacion = calcular_puntos(df_clasificacion)

        # Paso 7. Ordenar la tabla por  puntos. En caso de igualdad, por fifere
        df_clasificacion = ordenar_clasificacion(df_clasificacion)


        print(f"\n\nEl n√∫mero total de goles que ha marcado cada equipo es:\n {df_clasificacion[['goles_favor']].sort_values(by='goles_favor', ascending=False)}\n\n")

        print(f"El nombre del equipo m√°s goleador es:\n {df_clasificacion[['goles_favor']].sort_values(by='goles_favor', ascending=False).head(1)}\n\n")

        print(f"El nombre del equipo m√°s goleado es: \n {df_clasificacion[['goles_encontra']].sort_values(by='goles_encontra', ascending=False).head(1)}\n\n")

        print(f"La clasificaci√≥n global es:\n {df_clasificacion.to_string()}")
        # display(df_clasificacion) # no sale el resultado en la misma ventana del scroll
        
```
