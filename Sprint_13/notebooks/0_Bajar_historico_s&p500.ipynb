# 1.- S&P500 Bajar las 25 de mayor capitalización

 - Lista actualizada de las 25 mayores empresas por market cap
 - Descarga desde el año 2000
 - Guardado en CSV
 - Control de errores
 - Barra de progreso
 - Carpeta organizada

```python
import yfinance as yf
import pandas as pd
from datetime import datetime
from tqdm import tqdm
import os

# ---------------------------------------------------------
# 1. Lista de las 25 mayores empresas del S&P 500 por market cap
#    (Tickers oficiales de Yahoo Finance)
# ---------------------------------------------------------
sp500_top25 = {
    "Apple": "AAPL",
    "Microsoft": "MSFT",
    "Nvidia": "NVDA",
    "Amazon": "AMZN",
    "Meta": "META",
    "Alphabet Class A": "GOOGL",
    "Alphabet Class C": "GOOG",
    "Berkshire Hathaway": "BRK-B",
    "Eli Lilly": "LLY",
    "Broadcom": "AVGO",
    "Tesla": "TSLA",
    "JPMorgan": "JPM",
    "UnitedHealth": "UNH",
    "Visa": "V",
    "Exxon Mobil": "XOM",
    "Mastercard": "MA",
    "Johnson & Johnson": "JNJ",
    "Procter & Gamble": "PG",
    "Home Depot": "HD",
    "Costco": "COST",
    "Merck": "MRK",
    "Chevron": "CVX",
    "AbbVie": "ABBV",
    "PepsiCo": "PEP",
    "Walmart": "WMT"
}

# ---------------------------------------------------------
# 2. Parámetros de descarga años completos desde el 1-1-2000 al 31-12-2025
# ---------------------------------------------------------
start_date = "2000-01-01"
end_date = "2026-01-01"   # es exclusivo → se descarga hasta el día anterior a end.
# end_date = datetime.today().strftime("%Y-%m-%d")

output_folder = "SP500_Top25_data"
os.makedirs(output_folder, exist_ok=True)

# ---------------------------------------------------------
# 3. Descarga de datos
# ---------------------------------------------------------
print(f"Descargando histórico diario del S&P 500 (Top 25) desde {start_date} hasta {end_date}\n")

for name, ticker in tqdm(sp500_top25.items()):
    try:
        df = yf.download(ticker, start=start_date, end=end_date)

        if df.empty:
            print(f"⚠️ No se pudieron obtener datos de {name} ({ticker})")
            continue

        df.to_csv(f"{output_folder}/{name}_{ticker}.csv")
    except Exception as e:
        print(f"❌ Error descargando {name} ({ticker}): {e}")

print("\n✅ Descarga completada. Archivos guardados en la carpeta:", output_folder)
```
